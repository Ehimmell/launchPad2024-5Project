```python
nameVectors = pd.read_csv('professorNameVectors.csv')

names = nameVectors['names']

del nameVectors['names']
nameVectorsNP = nameVectors.values

NN = NearestNeighbors(n_neighbors=5)

NN.fit(nameVectorsNP)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
inputVector = model.encode('Andrew Freedman')

inputVector
```

```python
distances, indices = NN.kneighbors([inputVector])

indices[0]
```
